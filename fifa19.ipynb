# Clean Currency

```python
def clean_currency(val):
    if isinstance(val, str):
        val = val.replace('€', '')
        if 'M' in val:
            val = val.replace('M', '')
            val = float(val) * 1000000
    
        elif 'K' in val:
            val = val.replace('K', '')
            val = float(val) * 1000
                
    return(val)
```

```python
#fifa['Value in M€'] = fifa['Value'].apply(clean_currency)
fifa['Release Clause in M€'] = fifa['Release Clause'].apply(clean_currency)
fifa['Value in K€'] = (fifa['Value'].apply(clean_currency).astype(float)).astype(int)
fifa = fifa.drop(['Value', 'Release Clause'], axis=1)
fifa.head()
```

# Splitting dataset 

```python
X = fifa.drop('Value in K€', axis=1)
Y = fifa['Value in K€']
```
